> Interest on a bank loan is charged at fixed rate above the Bank of England 'official Bank Rate'.

> 1. Write a program that computes the interest payable, using the following variables in your program:
> 
>     1. Loan principal (amount borrowed)
>     1. Official Bank Rate (percentage, expressed per annum))
>     1. Rate over the official Bank Rate (percentage, expressed per annum)
>     1. Time in days
> 
>   Test your program with a loan principal of Â£150,000, bank rate of 0.25%, rate over the

```python
# Part IA Computing: Michaelmas Term
# Edwin Bahrami Balani (eb677@cam.ac.uk)

# Exercise 01.2
# Version A - no user input

principal = 150000
bank_rate = 0.25 / 100
rate_over_br = 1.49 / 100
period = 28

interest_rate = bank_rate + rate_over_br

interest = principal * interest_rate * (period / 365)

def p_tab(*args):
    print(*args, sep='\t')

p_tab("Principal:\t\t", 'Â£ '+str(principal))
p_tab("Official Bank Rate:\t", str(bank_rate*100)+'%')
p_tab("Rate over Official Bank Rate:", str(rate_over_br*100)+'%')
p_tab("Period:\t\t\t", str(period)+" days")
p_tab()
p_tab("Calculated interest rate:", str(interest_rate*100)+'%') # Float arithmetic...

p_tab("INTEREST PAYABLE = \t", 'Â£ '+str(interest))
```

## Explanation of the program
The *effective* interest rate of our loan equals $\text{official bank rate} + \text{rate over bank rate}$

Thus, $\text{interest rate} = 0.25\% + 1.49\% = 1.74\%$

However, while `interest_rate` is per annum, our `period` is expressed in days. We will assume that the interest is not compounded, i.e. interest is only payable on the principal, not on already-accumulated interest as well.

So, the interest payable in a given period of days is simply $\text{interest payable per year} \times \frac{\text{period in days}}{365}$.

$$
\text{interest payable per year} = \text{principal} \times \text{interest rate} \\
\therefore \text{interest payable} = \text{principal} \times \text{interest rate} \times \frac{\text{period in days}}{365}
$$

```python
# Part IA Computing: Michaelmas Term
# Edwin Bahrami Balani (eb677@cam.ac.uk)

# Exercise 01.2
# Version B - some user input

principal = float(input("Enter loan principal: "))
bank_rate = 0.25 / 100
rate_over_br = 1.49 / 100
period = int(input("Enter loan period in days: "))

interest_rate = bank_rate + rate_over_br

interest = principal * interest_rate * (period / 365)

def p_tab(*args):
    print(*args, sep='\t')

p_tab("Principal:\t\t", 'Â£ '+str(principal))
p_tab("Official Bank Rate:\t", str(bank_rate*100)+'%')
p_tab("Rate over Official Bank Rate:", str(rate_over_br*100)+'%')
p_tab("Period:\t\t\t", str(period)+" days")
p_tab()
p_tab("Calculated interest rate:", str(interest_rate*100)+'%') # Float arithmetic...

p_tab("INTEREST PAYABLE = \t", 'Â£ '+str(interest))
```
