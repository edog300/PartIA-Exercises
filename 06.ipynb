$$
A = \left| \frac{x_a(y_b  - y_c) + x_b(y_c - y_a) + x_c(y_a - y_b)}{2} \right|
$$

```python
# Part IA Computing: Michaelmas Term
# Edwin Bahrami Balani (eb677@cam.ac.uk)

# Exercise 06.1

# Two helper functions (see return statement in function)
x = lambda t: t[0]
y = lambda t: t[1]

# Side note: I was tempted to write this function using tuples in Activity 04 anyway...
def area_triangle(a, b, c):
    """
    Function that accepts three 2-D Cartesian (x, y) tuples
    a, b and c and returns the area of the triangle formed
    by them
    """
    for i in (a, b, c):
        if type(i) != tuple:
            print("Error!", i, "is not a tuple")
            return -1
        if len(i) != 2:
            print("Error!", i, "does not have 2 elements")
            return -1
        for j in i:
            if type(i) != int and type(i) != float:
                print("Error!", j, "is not an int or float")
                return -1
    # NB: a[0] is the x component of a i.e. x_a
    return abs((x(a)*(y(b)-y(c)) + x(b)*(y(c)-y(a)) + x(c)*(y(a)-y(b)))/2) #uses helper functions

# TODO add tests and demonstrations of the function
```

> ## Exercise 06.3 (indexing)
> 
> Write a function that uses list indexing to add two vectors of arbitrary length, and returns the new vector. Include a check that the vector sizes match, and print a warning message if there is a size mismatch. The more error information you provide, the easier it would be for someone using your function to debug their code.
> 
> Add some tests of your code.
> 
> *Hint*: You can create a list of zeros of length `n` by
> 
>     z = [0]*n
>     
> *Optional (advanced):* Try writing a one-line version of this operation using list comprehension and the built-in function `zip`.

```python
# Part IA Computing: Michaelmas Term
# Edwin Bahrami Balani (eb677@cam.ac.uk)

# Exercise 06.3

# Decorator to show the function being called
def show_call(func):
    def call_wrapper(*args):
        print("Calling", func.__name__+str(args))
        return func(*args)
    return call_wrapper

@show_call
def add_vectors(a, b):
    if type(a) not in (list, tuple) or type(b) not in (list, tuple):
        print("Error! a and b must be lists or tuples")
        return None
    
    for _ in (a, b):
        for i in _:
            if type(i) != int and type(i) != float:
                print("Error!", i, "is not an int or float", "("+str(type(i))+")")
                return None
            
    if len(a) != len(b):
        print("Warning! a and b are not the same size", "("+str(len(a)), "and", str(len(b))+")")
        print("Will only sum the first", min(len(a), len(b)), "elements of each")
    
    return [a[i]+b[i] for i in range(min(len(a), len(b)))]

# This is an experiment
# This function *should* be able to take an arbitrary number of vectors
@show_call
def add_vectors_alt(*vec):    
    return [sum(i) for i in zip(*vec)]

a = (3, 6, 7, 3, 8)
b = [2, 7, 3, 2]
print("Summing", a, b)
result1 = add_vectors(a, b)
print(result1)
print()

c = (7, 2, 43, 3, 6, 3, 6, 4, 7)
d = [1, 3, 1]
print("Summing", a, b, c, d)
result2 = add_vectors_alt(a, b, c, d)
print(result2)
print("Note how zip() silently only takes the first", min([len(i) for i in (a, b, c, d)]), "elements")
print()

@show_call
def interesting(*stuff):
    print("########################## STUFF ##########################")
    print("Demonstrating vec vs *vec:\n")
    print(stuff, "- note nested collections using stuff")
    print("i.e. equivalent to print((a, b) )")
    print()
    print(*stuff, "- note collections  have been 'expanded' with *vec")
    print("i.e. equivalent to print(a, b)")
    print("######################## END STUFF ########################")
print("(a, b) =", (a, b))
interesting(a, b)
```

Here, we want to pick the college name (from `numbers.keys()`) with the highest corresponding number of students. In the max function, the `key=` argument is a function argument. You pass to it a one-argument function that returns the "comparison key" - i.e. the value that is compared between items to determine the maximum value (and not to be confused with the dictionary key!!). See ["Built-in Functions"](https://docs.python.org/3/library/functions.html#max) in the Python documentation for more information.
